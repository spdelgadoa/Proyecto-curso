BUILD WITH MULTICRIERIA

Here we'll be modifying the DataFrame depending on some conditions & And |

```python
df_books.loc[(df_books['Is it good?'] =='Yes') & (df_books['Year']>2016), 'Name'] = 'Do watch'
df_books
```

```python
df_books['Watch?'] = np.where ((df_books['Is it good?'] =='Yes') & (df_books['Year']>2016), 'Yes','No')
df_books
```

CREATE A SUBSET TO CREATE DYNAMIC NEW COLUMNS

```python
df_subset = df_books[['Name', 'Author']]
new_cols = [str(x) + '_new' for x in df_subset]
categories = {1:'Alpha', 2:'Beta',3:"Gamma"}
df_books[new_cols] = df_books[df_subset].map(categories.get)
df_books
```
